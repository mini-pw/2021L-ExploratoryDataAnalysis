### Kod poprawionego wykresu

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {
    "Program": ["Dance Dance", "Twoja Twarz Brzmi Znajomo"],
    "rednio": [1.823, 1.543],
    "Pik": [2.023, 1.865]
}

df = pd.DataFrame(data)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)

fig.set_figheight(5)
fig.set_figwidth(15)

fig.suptitle('Porwnanie ogldalnoci dwch programw', size=15, x=0.5, y=0.99)


ax1.bar(df['Program'], df['rednio'], 0.7)
ax2.bar(df['Program'], df['Pik'], 0.7)


ax1.set_title('rednia ogldalno')
ax1.set_xlabel('Program')
ax1.set_ylabel('Ogldalno w mln')

ax2.set_title('Pikowa ogldalno')
ax2.set_xlabel('Program')
ax2.set_ylabel('Ogldalno w mln')


rects1 = ax1.bar(df['Program'], df['rednio'])
rects2 = ax2.bar(df['Program'], df['Pik'])

ax1.bar_label(rects1, padding=3)
ax2.bar_label(rects2, padding=3)

plt.style.use("seaborn")
```

### Uzasadnienie

Na poprawionym wykresie supki s pokazanie w 2D zamiast w 3D, co pozwala na lepsze porwnywanie
wysokoci supkw przez odbiorc. Porwnanie uatwiaj te linie siatki w tle. Dodatkowo, zakadajc, e wykres mia porwnywa dwa programy, a nie redni i pikow ogldalno dla tych programw, w nowym wykresie zmienione jest uoenie poszczeglnych kolumn. Po lewej stoj kolumny obu programw dla redniej ogldalnoci, a po prawej dla pikowej
ogldalnoci. Dziki temu moemy atwiej porwna przedstawione na wykresie programy.

## Wykres poprawiony i jego kod

```python



x = np.arange(len(dane))

width = 0.35

plt.figure(figsize=(10, 5))

plt.bar(
    x - width/2,
    dane['Men'],
    width,
    label='Male'
)

plt.bar(
    x + width/2,
    dane['Women'],
    width,
    label='Female'
)

ax = plt.gca()

ax.set_xticks(x)
ax.set_xticklabels(dane['Industry'])
plt.title('gender pay gap across U.S. industries in 2019')
plt.xlabel('Industry')
plt.ylabel('Yearly median income')
plt.xticks(rotation=90)

plt.legend()

plt.show()
```

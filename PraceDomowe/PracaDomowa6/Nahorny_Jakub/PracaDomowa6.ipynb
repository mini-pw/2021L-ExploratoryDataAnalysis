```python
data = {'Stopien':  ['(5+2)', '(5+1)', '(5+0)', '(4+2)', '(4+1)', '(4+0)', '(3+2)', '(2+2)', '(3+1)', '(3+0)', '(1+2)', '(2+1)'],
        'Udzial': [1, 5 , 13, 125, 1609, 2746, 4671, 60999, 63537, 108440, 316653, 838595]}
                     
Stopien = ['(5+2)', '(5+1)', '(5+0)', '(4+2)', '(4+1)', '(4+0)','(3+2)', '(2+2)', '(3+1)', '(3+0)', '(1+2)', '(2+1)']
Udzial = [1, 5 , 13, 125, 1609, 2746, 4671, 60999, 63537, 108440, 316653, 838595]
Wygrana = [63605427, 556722, 75573, 2620, 183, 83, 42, 17, 15, 13, 8, 7]
Pula = [0]*12
Procent = [0]*12
```

```python
for i in range(12):
    Pula[i] = Wygrana[i]*Udzial[i]

```

```python
for i in range(12):
    Procent[i] = Pula[i]/(sum(Pula))*100
```

```python
plt.bar(Stopien, Procent)
plt.xlabel('Kategorie')
plt.ylabel('Procent wygranych w kategorii')
plt.title('Eurojackpot podzia puli')
plt.grid(True)
plt.rcParams.update({'font.size': 4})



plt.show()
```

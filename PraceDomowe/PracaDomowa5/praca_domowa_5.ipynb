## Zadanie 3

Wyświetl nazwę użytkownika, który ma drugą największą liczbę odpowiedzi (również znajdujących się w ramce Posts) oznaczonych jako zaakceptowane rozwiązanie.

```python
acc = posts.copy(deep=True)
acc = acc[acc['Id'].isin(acc['AcceptedAnswerId'])]
acc['OwnerUserId'].value_counts()\
    .reset_index()\
    .merge(users[['Id','DisplayName']], left_on='index', right_on='Id')\
    .loc[1,'DisplayName']
```

## Zadanie 4

 Wyświetl odsetek postów, dla których odpowiedź zaakceptowana ma niższy wynik niż odpowiedź o najwyższym wyniku.

```python
answers = posts.copy(deep= True)
answers['Accepted'] = answers.Id.isin(answers.AcceptedAnswerId)
answers\
    .groupby('ParentId')\
    .apply(lambda x:np.nan if any(x.Accepted) == False\
           else (True if x.Score.max() > x.loc[x.Accepted].Score.max()\
           else False)).mean() * 100
```
